### DATA CLEANING

```python
#renaming some columns
df.rename(columns= {'Country/Region': 'Country', 'Electricity consumption (GW·h/yr)' : 'Electricity consumption'}, inplace = True)
```

```python
#changing the data type of Electricity consumption from object to float
df['Electricity consumption'] = pd.to_numeric(df['Electricity consumption'],errors = 'coerce')
```

```python
#dropping unimportant columns
df.drop(columns = ['As of','Year of data', 'Source'], inplace = True )
```

```python
#dropping the first row
df.drop([0], axis = 0, inplace = True )
df
```
